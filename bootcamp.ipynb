# Prologue 機械学習/AI と数学の関係

## Python を使って数学を学ぶ
### 機械学習との関連が深い、数学の４分野
- 確率・統計
- 微分・積分
- 線形代数
- 集合と位相

## 「機械学習の 3+1 Step」に見る数学との関係
### 機械学習Modelを利用する際の流れ
1. 機械学習Model を設計する
2. 誤差関数を定義する
3. 誤差関数を最小化するように Model を学習する
4. 学習済みModel の動作を確認する
   ※ 近年、XAI（説明可能な AI）という Keyword とともに、機械学習の **予測の理由** を説明するための技術も注目されている。

## 「線形代数」は機械学習Model の基本要素
機械学習の Model は、入力Data を受け取ったあと、一定の計算処理を行なって予測結果を出力する。== 数学の **「関数」** にあたる。
機械学習でさまざまな種類の関数を用いた Model が考案されているが、その中で最も基本となるのが **「一次関数」** を利用した **「線形Model」**。
この一次関数の性質を徹底的に調べるのが **「線形代数」** 。
線形代数の行列計算によって **「複数の一次関数をまとめて処理する」** という役割があり行列計算のによってスッキリ理解できるはず。

## 誤差関数に隠された「確率・統計」の考え方
機械学習Model の学習処理で必要になるのが **「誤差関数」** 。
機械学習Model の中には Tuning対象の Parameters が含まれる。Parameters の値を変化させて **「最も適切に分類できる」**ための Parameter値を見つけることが学習処理になる。
そのために、「何を持って適切に予測できているとするか」を決める基準を与えるのが誤差関数。
誤差関数の値が小さくなるほど、より高い精度で予測できることになるため、誤差関数の値がなるべく小さくなるように Tuning していく。
一般的に、Model が正しく予測する**「確率」**を計算して、この確率をもとに誤差関数を定義する。
機械学習で確率・統計を利用する場合は、誤差関数だけではないが、確率・統計を学ぶことで誤差関数のあたりから機械学習との関連が見えてくる。

## 微分計算で実現する勾配降下法
誤差関数が決まったら、勾配降下法で誤差関数の値を小さくするように Model の Parameters を Tuning する。
機械学習Library で Model の学習を行なう際、その内部では、**「微分計算で勾配Vector を求めたうえで、その反対方向に Parameters を修正する」** という処理が繰り返されている。
微積の知識があれば、**学習が進捗する様子を見ながら、その背後で行なわれている処理**が理解できる。

## 「集合と位相」で機械学習のしくみを理解する
学習が完了した後に Model の動作を分析するという作業がある。
近年の機械学習Model では、「潜在空間」に Data の特徴量を Mapping する、という処理がよく行なわれ潜在空間における Data の分布を分析することで、Model が学習した内容を調べることができる。
このような**空間上に散らばった Data の関連性を理解する**際は、集合と位相の考え方が役に立つ。

### 補足
上記で数学の４つの分野と機械学習の関わりを整理したが、あくまで一例に過ぎず。実際には機械学習の様々な場面で、さまざまな分野の数学が関連している。
